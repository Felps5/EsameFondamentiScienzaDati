```python
import os

import numpy as np
import pandas as pd
```

```python
path = os.path.join( 'Data', 'Raw' )
nome_file_tiri = 'NBA_2004_2024_Shots.csv'
nome_file_pbp = 'play_by_play.csv'
df_tiri = pd.read_csv( os.path.join( path, nome_file_tiri ) )
```

```python
df_pbp = pd.read_csv( os.path.join( path, nome_file_pbp ) )
```

```python
df_tiri[df_tiri.duplicated(keep=False)==True]
```

```python
df_tiri.drop_duplicates(inplace=True)
```

```python
df_tiri.columns
```

```python
df_tiri.dtypes
```

```python
df_tiri.head()
```

```python
df_tiri = df_tiri[df_tiri['SEASON_1']<2024]
```

```python
df_tiri['MIN_STR'] = df_tiri['MINS_LEFT'].map(str)
df_tiri['SEC_STR'] = df_tiri['SECS_LEFT'].map(str)
```

```python
df_tiri.dtypes
```

```python
df_tiri['SEC_STR'] = df_tiri['SEC_STR'].apply('{:0>2}'.format)
df_tiri['TIME_LEFT'] = df_tiri['MIN_STR'] + ':' + df_tiri['SEC_STR']
```

```python
df_tiri[df_tiri['POSITION'].isnull()]
```

```python
giocatori_senza_ruolo = pd.Series(df_tiri[df_tiri['POSITION'].isnull()]['PLAYER_NAME'].unique())
giocatori_senza_ruolo
```

```python
ruoli_mancanti = pd.Series(['PG', 'SG', 'SG', 'PF', 'PF', 'PF', 'PF', 'SG', 'SF', 'SF', 'PF', 'C', 'SG', 'SF', 'SF', 'PF', 'SG', 'SG', 'SG', 'C']) 
posgroup = ({'PG' : 'G', 'SG' : 'G', 'SF' : 'F', 'PF' : 'F', 'C': 'C'})
ruoli_mancanti = pd.DataFrame({'giocatore' : giocatori_senza_ruolo, 'ruolo' : ruoli_mancanti})
```

```python
n = int(ruoli_mancanti.size/2)
for i in range(0, n):
    df_tiri.loc[df_tiri.PLAYER_NAME == ruoli_mancanti.loc[i, 'giocatore'], 'POSITION'] = ruoli_mancanti.loc[i, 'ruolo']
    df_tiri.loc[df_tiri.PLAYER_NAME == ruoli_mancanti.loc[i, 'giocatore'], 'POSITION_GROUP'] = posgroup[ruoli_mancanti.loc[i, 'ruolo']]
```

```python
df_tiri[df_tiri['POSITION'].isnull()]
```

```python
df_tiri.isnull().sum()
```

```python
df_tiri.loc[(df_tiri['SHOT_DISTANCE'] == 0) | (df_tiri['SHOT_DISTANCE'] == 2), 'SHOT_DISTANCE'] = 1
```

```python
df_pbp.columns
```

```python
df_pbp.head()
```

```python
df_pbp['homedescription'] = df_pbp['homedescription'].fillna(df_pbp['visitordescription'])
```

```python
df_pbp.head(1000)
```

```python
df_pbp.drop(columns=['eventnum', 'eventmsgtype', 'eventmsgactiontype',
       'wctimestring', 'neutraldescription', 'visitordescription', 'score', 'scoremargin', 'person1type',
       'player1_name', 'player1_team_id', 'player1_team_city',
       'player1_team_nickname', 'player1_team_abbreviation', 'person2type',
       'player2_name', 'player2_team_id', 'player2_team_city',
       'player2_team_nickname', 'player2_team_abbreviation', 'person3type',
       'player3_name', 'player3_team_id', 'player3_team_city',
       'player3_team_nickname', 'player3_team_abbreviation',
       'video_available_flag'], inplace=True)
```

```python
df_pbp['ASSISTED'] = df_pbp['homedescription'].str.contains('AST')
```

```python
df_pbp.head(5)
```

```python
df_tiri_merged = pd.merge(df_tiri, df_pbp, left_on=['GAME_ID', 'QUARTER', 'TIME_LEFT'], 
                          right_on=['game_id', 'period', 'pctimestring'], how='left')
df_tiri_merged.drop(columns=['game_id', 'period', 'pctimestring',
       'homedescription', 'player1_id', 'player2_id', 'player3_id'], inplace=True)
df_tiri_merged.dropna(inplace=True)
```

```python
df_tiri_merged.to_parquet( os.path.join('Data','Parquet','nbashots.parquet') )
```
