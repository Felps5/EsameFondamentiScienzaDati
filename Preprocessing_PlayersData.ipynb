```python
import os
import numpy as np
import pandas as pd
```

```python
#Carico in un dataframe il file game.csv contente i dati su tutte le partite dal 1946 al 2023
path = os.path.join('Data','Raw')
nome_file = 'player_totals.csv'

df = pd.read_csv( os.path.join( path, nome_file ) )
file_teams = 'Team Abbrev.csv'
teams_abb = pd.read_csv( os.path.join(path, file_teams))
#Visualizzo i nomi delle colonne per facilitare i passi successivi
df.columns
```

```python
df['pos'].unique()
```

```python
teams_abb.drop_duplicates('abbreviation', inplace=True)
teams_abb = teams_abb[teams_abb.season>1985]
teams_abb
```

```python
df = pd.merge(df, teams_abb, left_on='tm', right_on='abbreviation')
df.drop(['season_y', 'lg_y', 'playoffs', 'abbreviation'], axis='columns', inplace=True)
df
```

```python
df.rename(columns={'season_x' : 'season'}, inplace=True)
```

```python
df.team.unique()
```

```python
df.loc[df.team=='Charlotte Bobcats' , 'tm'] = 'CHO'
df.loc[df.team=='New Orleans Hornets', 'tm'] = 'NOP'
df.loc[df.team=='New Jersey Nets', 'tm'] = 'BRK'
df.loc[df.team=='Seattle SuperSonics', 'tm'] = 'OKC'
df.loc[df.team=='New Orleans/Oklahoma City Hornets', 'tm'] = 'CHO'
df.loc[df.team=='Vancouver Grizzlies', 'tm'] = 'MEM'
df.loc[df.team=='Washington Bullets', 'tm'] = 'WAS'
```

```python
df.loc[df.team=='Charlotte Bobcats' , 'team'] = 'Charlotte Hornets'
df.loc[df.team=='New Orleans Hornets', 'team'] = 'New Orleans Pelicans'
df.loc[df.team=='New Jersey Nets', 'team'] = 'Brooklyn Nets'
df.loc[df.team=='Seattle SuperSonics', 'team'] = 'Oklahoma City Thunder'
df.loc[df.team=='New Orleans/Oklahoma City Hornets', 'team'] = 'Charlotte Hornets'
df.loc[df.team=='Vancouver Grizzlies', 'team'] = 'Memphis Grizzlies'
df.loc[df.team=='Washington Bullets', 'team'] = 'Washington Wizards'
```

```python
stats = ['mp', 'fga', 'fta', 'tov']
teams_totals = df.groupby(['season', 'tm'])[['mp', 'fga', 'fta', 'tov']].sum().reset_index()
teams_totals.rename(columns={'mp' : 'team_mp', 'fga' : 'team_fga', 'fta' : 'team_fta', 'tov' : 'team_tov'}, inplace=True)
df = pd.merge(df, teams_totals, on=['season', 'tm'], how='left')
df
```

```python
df['player_usage'] = 100 * ( (df.fga + 0.44*df.fta + df.tov) * (df.team_mp/5) ) / ( df.mp * (df.team_fga + 0.44*df.team_fta + df.team_tov)) 
df['player_usage']
```
