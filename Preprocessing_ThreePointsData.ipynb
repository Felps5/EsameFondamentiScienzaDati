```python
import os
import numpy as np
import pandas as pd
```

```python
#Carico in un dataframe il file game.csv contente i dati su tutte le partite dal 1946 al 2023
path = os.path.join('Data','Raw')
nome_file = 'game.csv'

df = pd.read_csv( os.path.join( path, nome_file ) )

#Visualizzo i nomi delle colonne per facilitare i passi successivi
df.columns
```

```python
#Esploro il dataframe
df
```

```python
#Dalle ultime righe si nota la presenza di dati duplicati, elimino dunque i duplicati basando la ricerca sulla colonna 'game_id'
df.drop_duplicates('game_id', inplace=True)
df
```

```python
#Controllo quali sono tutti i possibili valori della colonna 'season_type', che indica a quale tipo di partita di riferisce la relativa osservazione
df['season_type'].unique()
```

```python
#Inserisco una colonna con l'anno in cui si sono disputate le partite
df.insert(0, 'year', df['season_id'].apply(str))
df['year'] = df['year'].str[1:].astype(int)
df['year'] += 1
```

```python
#Il tiro da 3 è stato aggiunto al regolamento dal 1979, escludo quindi i dati relativi alle stagioni precedenti.
#Escludo inoltre i dati relativi a tutti i match non validi per il campionato, quindi le amichevoli Pre-season e le esibizioni degli All Star
df = df[(df['year']>=1979) & ((df['season_type']=='Regular Season') | (df['season_type']=='Playoffs'))]

df
```

```python
#Le prime righe presentano valori NaN, conto i valori per verificare in quali annate si verifica questa evenienza
df.groupby('year').count()

#% partite playoff su totale per giustificare eliminazione dati 2012
df[ (df['year']==2022) & (df['season_type']=='Playoffs') ].count().iloc[1]/(df[ (df['year']==2022) & (df['season_type']=='Playoffs') ].count().iloc[1]+df[ (df['year']==2022) & (df['season_type']=='Regular Season') ].count().iloc[1])*100
```

```python
#Osservando il conteggio dei valori si nota che fino al 1985 incluso vi sono dei problemi nella rilevazione di alcuni dati,
#in particolare per quanto concerne il numero di tiri da 3 tentati dalle squadre. Mentre per il 2012 mancano i dati della stagione regolare.
#Considero solo le stagioni in cui ho le osservazioni complete.
df = df[ ( df['year'] >= 1986 ) & ( df['year'] != 2013) ]

#Verifico di non trovare altri valori nulli
df.isnull().sum()
```

```python
df.groupby('year').count()
```

```python
# Inserisco le colonne relative al tiro da 2 punti per futuri calcoli
df.insert(7, 'fg2m_home', (df['fgm_home']-df['fg3m_home']))
df.insert(8, 'fg2a_home', (df['fga_home']-df['fg3a_home']))
df.insert(39, 'fg2m_away', (df['fgm_away']-df['fg3m_away']))
df.insert(40, 'fg2a_away', (df['fga_away']-df['fg3a_away']))
```

```python
#A questo punto reimposto il contatore delle righe del dataframe
df.reset_index(inplace=True)
df = df.loc[ :, 'year': ]

#Salvo i dati ricavati in formato Parquet per successive analisi
df.to_parquet( os.path.join('Data','Parquet','threepoints.parquet') )
```
