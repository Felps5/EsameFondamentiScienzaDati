```python
import os
import numpy as np
import pandas as pd
```

```python
path = os.path.join('Data','Raw')
nome_file = os.listdir( path )
```

```python
for f in nome_file:
    df = pd.read_csv( os.path.join( path, f ) )
    print('Nome file: ', f)
    print('Colonne: ')
    for c in df.columns:
        print("- ", c)
    print("\n")
```
