```python
import os
import numpy as np
import pandas as pd
```

```python
#Carico in un dataframe il file game.csv contente i dati su tutte le partite dal 1946 al 2023
path = os.path.join('Data','Raw')
nome_file = 'game.csv'

df = pd.read_csv( os.path.join( path, nome_file ) )

df.insert(0, 'year', df['game_date'].str[:4].astype(int))

#Visualizzo i nomi delle colonne per facilitare i passi successivi
df.columns
```

```python
#Esploro il dataframe
df[(df['year']==1998) & (df['fg3a_home']==0)]
```

```python
#Dalle ultime righe si nota la presenza di dati duplicati, elimino dunque i duplicati basando la ricerca sulla colonna 'game_id'
df.drop_duplicates('game_id', inplace=True)
df
```

```python
#Controllo quali sono tutti i possibili valori della colonna 'season_type', che indica a quale tipo di partita di riferisce la relativa osservazione
tipo_match = df['season_type'].drop_duplicates()
tipo_match
```

```python
#Inserisco una colonna con l'anno in cui si sono disputate le partite
```

```python
#Il tiro da 3 è stato aggiunto al regolamento dal 1979, escludo quindi i dati relativi alle stagioni precedenti.
#Escludo inoltre i dati relativi a tutti i match non validi per il campionato, quindi le amichevoli Pre-season e le esibizioni degli All Star
df = df[(df['year']>=1979) & ((df['season_type']=='Regular Season') | (df['season_type']=='Playoffs'))]

#Considero solo le colonne utili alle considerazioni sull'analisi dei tiri da 3 per annata
df = df.loc[:,['year','fg3m_home','fg3a_home','fg3m_away','fg3a_away']]
df
```

```python
#Le prime righe presentano valori NaN, verifico per quante annate si verifica questa evenienza
df.groupby('year').count()
```

```python
#Osservando il conteggio dei valori si nota che fino al 1985 incluso vi sono dei problemi nella rilevazione di alcuni dati,
#in particolare per quanto concerne il numero di tiri da 3 tentati dalle squadre. Considero solo le stagionii in cui ho le osservazioni complete
df = df[(df['year']>=1986)]

#Verifico di non trovare altri valori nulli
df.isnull().sum()
```

```python
#A questo punto reimposto il contatore delle righe del dataframe e 
df.reset_index(inplace=True)
df = df.loc[:,'year':]

#Rinomino le colonne con nomi più intellegibili
df.rename(columns={'fg3m_home' : 'made_home', 'fg3a_home' : 'attempted_home',	'fg3m_away' : 'made_away', 'fg3a_away' : 'attempted_away'}, inplace=True)

#Salvo i dati ricavati in formato Parquet per successive analisi
df.to_parquet( os.path.join('Data','Parquet','threePointsTotal.parquet') )
```
